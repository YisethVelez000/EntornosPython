```python
import pandas as pd
import requests


url = 'https://dummyjson.com/products'
response = requests.get(url)

if response.status_code== 200:
    data = response.json()
    df= pd.DataFrame(data['products'])
    
    
```

Generar una grafica de cantidad de productos por marca

```python
from dash import Dash, html, dash_table,dcc
import plotly.express as px 
import dash_bootstrap_components as dbc 



# Iniciamos la app
app = Dash(__name__,external_stylesheets=[dbc.themes.BOOTSTRAP])
app.layout = html.Div([
    dbc.Row(
        [
            dbc.Col(html.H2("Productos"), width=12)
        ]
    ),
    
# Generamos la grafica de la cantidad de prouctos, su titulo y el stock
    dbc.Row([
        dbc.Col(dcc.Graph(figure=px.bar(df, x='title', y='stock', title='Cantidad de productos')), width=6),
    ]),

#Generamos la grafica de la cantidad de productos por marca
    dbc.Row([
        dbc.Col(dcc.Graph(figure=px.pie(df, names='brand', title='Cantidad de productos por marca')), width=6),
    ]),
    
    
])

# Corremos la app 
if __name__ == '__main__':
    app.run(debug=True)
```

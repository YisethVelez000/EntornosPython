```python
#Importamos las librerias
import PyPDF2
import openpyxl
```

```python
#Creamos el archivo de excel
wb = openpyxl.Workbook()

#Creamos la hoja de trabajo
ws = wb.active
```

```python
#Escribimos los encabezados
ws['A1'] = "Nombre Completo"
ws['B1'] = 'Edad'
ws['C1'] = 'Ciudad de Colombia'
ws['D1'] = 'Tiempo Corrido'
ws['E1'] = 'Puesto'
```

```python
#Abrimos y recorremos el pdf

with open('Datos.pdf','rb') as pdf_file:
    #Creamos el objeto que lee el pdf
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    #Obtenemos el número de páginas 
    numeroPaginas = len(pdf_reader.pages)

    #Iteramos sobre las páginas y extraemos el texto 
    for pagina in range(numeroPaginas):
        page = pdf_reader.pages[pagina]
        text = page.extract_text()
        matriz = text.split()
        
        #Iteramos sobre la matriz y extraemos los datos los datos estan en el siguiente orden 
        #Nombre, Edad, Ciudad, Tiempo, Puesto los datos se encuentran en grupos de 5, por lo que iteramos de 5 en 5
        #La primera fila ya esta ocupada por los encabezados por lo que iniciamos en la fila 2, por eso el range(1,len(matriz),5)
        #El primer dato es Juan, por lo que debemos guardar desde ahi 
        for i in range(1,len(matriz),6):
            if i+5 < len(matriz):
                ws.append([matriz[i],matriz[i+1],matriz[i+2],matriz[i+3],matriz[i+4]])
                print(matriz[i],matriz[i+1],matriz[i+2],matriz[i+3],matriz[i+4])
            
            
            
                 
        
                

#Guardamos los datos
wb.save('datos.xlsx')
            
            
                  
#Guardamos los datos wb.save('datos.xlsx')
```
